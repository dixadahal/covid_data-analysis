```python
country_vaccinations_df=pd.read_csv('2021-05-19_country_vaccinations.csv')
country_vaccinations_by_manufacturer_df=pd.read_csv('2021-05-19_country_vaccinations_by_manufacturer.csv')
cv_df=country_vaccinations_df
cvm_df=country_vaccinations_by_manufacturer_df
```

```python
country_vaccinations_df.info()
```

```python
cv_df['date']=pd.to_datetime(cv_df['date'],format='%Y-%m-%d')
cvm_df['date']=pd.to_datetime(cvm_df['date'],format='%Y-%m-%d')
cv_df
```

```python
cv_df_grp = cv_df.groupby('country').max().reset_index()
cv_df_grp
```

### This figure below gives the map for the total vaccination for each countries. Quoting Group 1 in this figure
### "We find that China, India and United States some of the countries that has made the best vaccination progress."

### This figure gives the total vaccination done by each countries but this alone can not be determine the best vaccination progress.
### The population has to come into the mix for proper picture.

```python
figure = px.choropleth(cv_df_grp,
                       locations='iso_code', 
                       color='total_vaccinations', 
                       hover_name='country', 
                       color_continuous_scale='portland', 
                       title='Total COVID-19 Vaccinations',
                       featureidkey="properties.country",
#                        animation_frame='dt_str'
                      )

figure.update_layout(
    title_text = 'Total COVID-19 Vaccinations',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
        projection_type = 'equirectangular'
    ))
    
figure.show()
```

```python
### In this figure we can see the vaccination progress with respect to population of the country.
### The previous and this figure gives the information on how total vaccination does not necessarily mean good progress.
### here we can see Chile, Mongolia, UK are doing pretty good which cannot be seen in previous figure.
```

```python
figure = px.choropleth(cv_df_grp,
                       locations='iso_code', 
                       color='total_vaccinations_per_hundred', 
                       hover_name='country', 
                       color_continuous_scale='portland', 
                       title='Total COVID-19 Vaccinations',
                       featureidkey="properties.country",
#                        animation_frame='dt_str'
                      )

figure.update_layout(
    title_text = 'Total COVID-19 Vaccinations',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
        projection_type = 'equirectangular'
    ))
    
figure.show()
```

### The other metrics we can use to see vaccination progress is the difference between fully vaccinated and vaccinated with at least one dose. From these two figures we can see that chile has around 8 million people vaccinated where 7 million are fully vaccinated. Whereas India has 139 Million people vaccinated with only 39 Million fully vaccinated.  Group 1 has shown this relationship in a bar chart which does not cover most of the countries. 

```python
vac_df = cv_df.groupby(['date','iso_code','country']).sum().reset_index().sort_values('date', ascending=True).dropna()
mask = (vac_df['date'] > '2021-01-01') & (vac_df['date'] <= '2021-12-31')
vac_df['dt_str'] = vac_df['date'].apply(lambda x: x.strftime("%m/%d/%Y"))
timeline_df = vac_df.loc[mask]
timeline_df = timeline_df[timeline_df['daily_vaccinations']>0]
timeline_df
```

```python
figure = px.choropleth(cv_df_grp,
                       locations='iso_code', 
                       color='people_fully_vaccinated', 
                       hover_name='country', 
                       color_continuous_scale=px.colors.sequential.Turbo, 
                       title='People fully vaccinated by COVID-19 Vaccines')

figure.update_layout(
    title_text = 'People Fully Vaccinated by COVID-19 Vaccine',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
        projection_type = 'equirectangular'
    ))

figure.show()
```
